7. Add a BMI column: BMI = Weight[kg] / (Height[m])**2

```python
def lbs_to_kg(x):
  return x*0.453

def inches_to_m(y):
  return y*0.0254

df['Weight'].map(lbs_to_kg)
df['Height'].map(inches_to_m)
df['BMI'] = df['Weight']/(df['Height']**2)
print(df['BMI'])
df.head(20)
```

8. define categorical BMI classes

```python
df['cBMI']=pd.cut(df['BMI'],bins=[0,20,27,100],labels=['underweight','normal','overweight'])
```

8.1. groupby cBMI and compute average statistics

```python
df_cBMI = df.groupby(['cBMI'])
df_cBMI.head(20)
```
